```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Pretpostavimo da imamo cijene za 2021. godinu za tri operatora
operators = ['Deutsche Telekom', 'Vodafone', 'Telefonica (O2)']
prices_2021 = [30, 25, 35]  # Cijene za 2021. godinu (simulirani podaci)

# Kreiramo odgovarajui niz godina
years = np.array([2021, 2022, 2023, 2024, 2025]).reshape(-1, 1)

# Za sada samo simuliramo cijene koje rastu svake godine za 5% za svaki operator
price_increase_rate = 0.05
prices_2021_growth = []

# Kreiramo cijene za svaku godinu
for price in prices_2021:
    growth_prices = [price * (1 + price_increase_rate)**(year - 2021) for year in years.flatten()]
    prices_2021_growth.append(growth_prices)

# Kreiramo linearni model za svaki operator
model = LinearRegression()

# Predvianje cijena za 2025. godinu
predicted_prices_2025 = []

for operator_prices in prices_2021_growth:
    model.fit(years, operator_prices)  # Fit model za svakog operatora
    predicted_prices_2025.append(model.predict([[2025]])[0])

# Ispis rezultata
print(f"Predviene cijene za 2025. godinu: {predicted_prices_2025}")

# Vizualizacija
plt.figure(figsize=(10, 6))

# Crtanje cijena za svakog operatora
for i, operator in enumerate(operators):
    plt.plot(years.flatten(), prices_2021_growth[i], marker='o', label=f'{operator} - 2021-2025')

plt.scatter(2025, predicted_prices_2025, color='red', label='Predviene cijene za 2025. godinu', zorder=5)

plt.title('Predvianje cijena za 2025. godinu', fontsize=14)
plt.xlabel('Godina', fontsize=12)
plt.ylabel('Cijena (EUR)', fontsize=12)
plt.legend()
plt.grid(True)
plt.show()
```
